```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

# Dane wejściowe
house_price = 120000  # Obecna cena mieszkania w zł
rate_house = 0.05  # Roczny wzrost ceny mieszkania
periods = 5  # Okres oszczędzania w latach
rate_deposit= 0.12  # Nominalna stopa procentowa w skali roku
freq = 12  # Miesięczna kapitalizacja


monthly_rate_house = rate_house / freq  # konwersja stopy do okresu miesięcznego
monthly_rate_deposit = rate_deposit / freq
nper = periods * freq

periods = np.arange(1,nper+1,dtype=int)



```

```python
# 1)Ile będzie wynosiła orientacyjna cena mieszkania za 5 lat?
import numpy_financial as npf
new_house_price = np.around(npf.fv(monthly_rate_house, periods, pmt=0, pv=-house_price), 2)

future_price = new_house_price[-1]
print("Cena mieszkania za 5 lat będzie wynosiła", future_price, "zł")
new_house_price
```

```python
#Ile musisz wpłacać do banku każdego miesiąca, aby przy przedstawionej ofercie uzbierać na mieszkanie w ciągu 5 lat?
#npf.pmt(rate, nper, pv, fv, when='end')

#   rate – stopa procentowa dla jednego okresu (np. miesięczna stopa oprocentowania)
#   nper – liczba okresów (np. liczba miesięcy oszczędzania)
#   pv – wartość początkowa (tu 0, bo zaczynamy od zera)
#   fv – przyszła wartość (np. cena mieszkania, wpisujemy jako wartość dodatnią)
#   when – 'end' (wpłaty na koniec miesiąca, domyślnie) lub 'begin' (wpłaty na początku miesiąca)

```

```python
pv = 0
savings_calculation = np.around( npf.pmt(monthly_rate_deposit, periods, pv , -future_price, when='begin'), 2)
final_savings = savings_calculation[-1]

print("Musisz oszczędzać ",final_savings," zł miesięcznie")

savings_increase_by_month_look = - npf.fv(monthly_rate_deposit, periods, final_savings, final_savings)


```

```python
#Stwórz wykres przedstawiający, jak w interwałach miesięcznych zmieniać się będzie cena mieszkania (liniowy wzrost w całym okresie) oraz 
#wartość twojej lokaty

import matplotlib.pyplot as plt

plt.plot(new_house_price,label='cena mieszkania')
plt.plot(periods, savings_increase_by_month_look,label='oszczędności')
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Skumulowana wartość odsetek/mieszkania')
```
